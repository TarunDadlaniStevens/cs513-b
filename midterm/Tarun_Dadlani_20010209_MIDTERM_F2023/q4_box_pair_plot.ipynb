```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mplot
import seaborn as sns
```

```python
dataset = pd.read_csv('BP.csv')
dataset.head()
```

```python
dataset.info()
```

```python
dataset = dataset.replace("?", np.NaN)
dataset.isnull().sum()
```

```python
age_weight_height_plot_data = dataset.iloc[:, 2:5]
age_weight_height_plot_data.plot(kind='box', subplots=True, layout=(1, 3), figsize=(15, 5))
mplot.suptitle("Box Plots for Columns Age, Height and Weight")
mplot.show()
```

```python
sns.pairplot(age_weight_height_plot_data)
mplot.suptitle('Pair Plot of Age, Height, and Weight', y=1.02)
mplot.show()
```
